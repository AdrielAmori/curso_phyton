```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Criando nosso dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#visualizando as 5 primeiras linhas
df.head()
```

```python
#quantidade de linhas e colunas
df.shape
```

```python
#verificando os tipos de dados
df.dtypes
```

```python
#qual a receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
#Qual o custo total?
round(df["Custo"].sum(), 2)
```

```python
#Lucro total = Rceita - Custo
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
round(df["Lucro"].sum() ,2)
```

```python
#Criando uma coluna com total de dias para envio
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#verificando o tipo da coluna Tempo-envio
df["Tempo_envio"].dtypes
```

```python
#Média do tempo de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```

# Missing Values

```python
#verificando se temos dados faltantes
df.isnull().sum()
```

# Se quiser saber o Lucro por Ano e por Marca

```python
#Vamos agrupar pro ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Rsetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title = "Total produtos vendidos", color = "orangered")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#grafico lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().sort_values().plot.bar(title = "Lucro X Ano")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#selecionando as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title = "Lucro X Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title = "Lucro X Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```

```python
#Lucro por classe
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title = "Lucro X Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```
